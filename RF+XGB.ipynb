```python
df_train=pd.read_csv('train_final.csv')
df_test=pd.read_csv('test_final.csv')
```

```python
print(df_train.shape)
target=df_train.Y
df_train.drop('Y',axis='columns',inplace=True)
```

```python
frames=[df_train,df_test]
df=pd.concat(frames)
```

```python
ID=df.Id
df1=df.drop('Id',axis='columns')
test_ID=ID[16383:]
test_ID.shape
```

```python
Norm_df_train=Norm_df[:16383]
shuffle(Norm_df_train, random_state=1)
```

```python
Norm_df_test=Norm_df[16383:]
Norm_df_test
```

```python
from sklearn.model_selection import cross_val_score
alphas = [500] 
for i in alphas:
    model_randomforest= RandomForestClassifier(n_estimators=i,
                                               max_features='auto',
                                               min_samples_leaf=2,
                                               min_samples_split=2,
                                               max_depth=50,
                                               min_weight_fraction_leaf=0,
                                               max_leaf_nodes=1000,
                                               criterion='entropy',
                                               min_impurity_decrease=0,
                                               n_jobs=1,
                                               class_weight='balanced_subsample',
                                               ccp_alpha=0,
                                               random_state=0)
    n_scores = cross_val_score(model_randomforest, Norm_df_train, target, scoring='roc_auc', cv=5).mean()
    print(n_scores)
```

```python
sub_model=RandomForestClassifier(n_estimators=500,
                                               max_features='auto',
                                               min_samples_leaf=2,
                                               min_samples_split=2,
                                               max_depth=50,
                                               min_weight_fraction_leaf=0,
                                               max_leaf_nodes=1000,
                                               criterion='entropy',
                                               min_impurity_decrease=0,
                                               n_jobs=1,
                                               class_weight='balanced_subsample',
                                               ccp_alpha=0,
                                               random_state=0)
sub_model.fit(Norm_df_train,target)
```

```python
h=sub_model.predict_proba(Norm_df_test)
y_newpred=h[:,1]
y_newpred
```
